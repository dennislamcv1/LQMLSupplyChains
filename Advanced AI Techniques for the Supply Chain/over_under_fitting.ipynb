In this notebook we will examine the difficulties arising from under-fitting and over-fitting a model.

```python
### In this cell, we import the necessary packages.

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

%matplotlib inline
```

```python
### Generate the data that we want to fit.

x_data = np.linspace(0, 5, 20)
y_data = x_data**3 + 50*np.random.rand(20)
```

```python
### Create a scatter plot of the data.

plt.scatter(x_data, y_data)
plt.show()
```

```python
### Convert x_data into a matrix version to be used in the model.

x_mat = x_data[:, np.newaxis]
```

```python
### Perform a line of best fit and display the results.
### See how the model is too simple and underfits.

model = make_pipeline(PolynomialFeatures(1), Ridge())
model.fit(x_mat, y_data)

y_fit = model.predict(x_mat)

plt.scatter(x_data, y_data)
plt.plot(x_data, y_fit)
plt.show()
```

```python
### There are 20 data points so let's perform a polynomial of best fit of degree 19.
### See how the model is too complicated and overfits.

model = make_pipeline(PolynomialFeatures(19), Ridge())
model.fit(x_mat, y_data)

y_fit = model.predict(x_mat)

plt.scatter(x_data, y_data)
plt.plot(x_data, y_fit)
plt.show()
```

```python
### The original data was based on a cubic, so let's do a polynomial fit of degree 3.
### See how the model cuts right through the noise of the data nicely.

model = make_pipeline(PolynomialFeatures(3), Ridge())
model.fit(x_mat, y_data)

y_fit = model.predict(x_mat)

plt.scatter(x_data, y_data)
plt.plot(x_data, y_fit)
plt.show()
```
